```python
import fitz
import re
from transformers import AutoTokenizer
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text("text")+"\n"
    return text.strip()


pdf_text = extract_text_from_pdf('articel.pdf')
print(pdf_text)
```

```python
def clean_sinhala_contract_text(text):
    
    text = re.sub(r'[^A-Za-z0-9\u0D80-\u0DFF.,?!\s\-\'"(){}:;]', ' ', text)

    text = re.sub(r'\s+', ' ', text).strip()
    return text


pdf_text_cleaned = clean_sinhala_contract_text(pdf_text)
print(pdf_text_cleaned)
```

```python
chunk_size = 400


contract_chunks = [pdf_text_cleaned[i:i + chunk_size] for i in range(0, len(pdf_text_cleaned), chunk_size)]


contract_chunks = contract_chunks[:-1] if len(contract_chunks[-1]) < chunk_size else contract_chunks

print(f"Total Chunks: {len(contract_chunks)}")
```

```python
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer
```

```python
model = SentenceTransformer('Ransaka/bert-small-sentence-transformer')
```

```python
from transformers import AutoModel
```

```python
import torch
```

```python
import faiss
import numpy as np
```

```python
contract_embeddings = model.encode(contract_chunks, convert_to_numpy=True)
```

```python
from annoy import AnnoyIndex
```

```python
embedding_dim = contract_embeddings.shape[1]  # Get embedding size
annoy_index = AnnoyIndex(embedding_dim, 'angular')
```

```python
for i, emb in enumerate(contract_embeddings):
    annoy_index.add_item(i, emb)

annoy_index.build(20) 
```

```python
def get_answer(question, top_n=3):  
    question_embedding = model.encode([question], convert_to_numpy=True)[0]
    
  
    indices = annoy_index.get_nns_by_vector(question_embedding, top_n)
    

    answers = [contract_chunks[i] for i in indices]
    
    return answers
```

```python
question = "දණ්ඩ නීති සංග්‍රහයේ 345 වගන්තියට එක්වුණු නව සංශෝධන "
top_n = 1
answers = get_answer(question, top_n)

print("Top-N Answers:")
for idx, answer in enumerate(answers, 3):
    print(f"Answer {idx}: {answer}")
```
